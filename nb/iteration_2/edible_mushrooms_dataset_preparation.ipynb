```python
# creation d'un dataframe 'df_species_com' ne contenant que les champignon comestibles
df_species_com = df.loc[df['gbif_info/species'].isin(list_champi_com)]
```

```python
# il semble que la colonne 'gbif_info/species' soit le plu scorrel√©e avec la liste des champignons comestibles
# Nous choissirons de conserver cette variable comme variable cible.

# Exploitation et nettoyage du dataframe 'df_species_com'
df_species_com = df_species_com.drop(['gbif_info', 'gbif_info/kingdom', 'gbif_info/speciesKey','gbif_info/orderKey',
                                      'gbif_info/familyKey','gbif_info/kingdomKey','gbif_info/genusKey',
                                      'gbif_info/phylumKey','gbif_info/genusKey',
                                      'gbif_info/usageKey','gbif_info/classKey',
                                      'gbif_info/note','location', 'thumbnail', 
                                      'observation', 'image_url', 'user', 'date', 'label'], axis = 1)
```

```python
# suppression des NaNs
df_species_com = df_species_com.dropna(axis = 0)
```

```python
# Define a function to remove the prefix 'gbif_info' from column names
def remove_prefix(col):
    if col.startswith('gbif_info/'):
        return col[len('gbif_info/'):]
    return col

# Rename columns by removing the prefix 'gbif_info'
df_species_com = df_species_com.rename(columns=remove_prefix)

df_species_com.info()
```

# Structuring images

```python
src_path = '../../raw/'
dest_path = '../../dataset/'
```

```python
# sorting out images according to their classes
from PIL import Image
import shutil
import random
import os

test_size = 0.15

def random_sort_image(img_id, class_id, class_type):
    # keeping same file structure
    test_path = os.path.join(dest_path, class_type, 'test')
    train_path = os.path.join(dest_path, class_type, 'train')
    
    # sourcing image
    subdir = '0' if (img_id < 100000) else str(img_id)[0]
    src = os.path.join(src_path,  subdir, str(img_id) + '.jpg')

    # copying image in its class folder randomly choosing between training and testing
    if random.random() < test_size:
        dst = os.path.join(test_path, class_id, str(img_id) + '.jpg')
    else:
        dst = os.path.join(train_path, class_id, str(img_id) + '.jpg')
    
    try:
        shutil.copy(src, dst)
        # eg. src and dest are the same file
    except shutil.Error as e:
        print('Error: %s src=%s dest=%s' % (e, src, dst))        
        # eg. source or destination doesn't exist
    except IOError as e:
        print('Error: %s src=%s dest=%s' % (e.strerror, src, dst))   
```

```python
def create_subdirectories(df, column, parent_directory):
    # Get the unique values of the column
    unique_values = df[column].unique()
    
    # creating parent directories
    os.makedirs(parent_directory, exist_ok=True)
    
    # Create a subdirectory for each unique value
    for value in unique_values:
        subdir_path = os.path.join(parent_directory, str(value))
        os.makedirs(subdir_path, exist_ok=True)
```

```python
import os
import shutil

def remove_files_and_directories(directory, remove_directories=False):
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        try:
            if os.path.isfile(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path) and remove_directories:
                shutil.rmtree(file_path)
        except Exception as e:
            print(f'Failed to delete {file_path}. Reason: {e}')
```

```python
import os
# dataset >> class_type >> train, test >> class_id >> image

# preparing datasets for classification
for class_type in ['mixed']:
    create_subdirectories(df_species_com, class_type, os.path.join(dest_path, class_type, 'test'))
    create_subdirectories(df_species_com, class_type, os.path.join(dest_path, class_type, 'train'))
    df_species_com.loc[:, ['image_id',class_type]].apply(lambda x: random_sort_image(x['image_id'], x[class_type], class_type), axis=1)
```
